```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
url = "https://raw.github.com/mattdelhey/kaggle-titanic/master/Data/train.csv"
titanic = pd.read_csv(url)
titanic.info()
```

```python

titanic.head()
```

```python
def woman_child_or_man(passenger):
    age, sex = passenger
    if age < 16:
        return "child"
    else:
        return dict(male="man", female="woman")[sex]
```

```python
titanic["class"] = titanic.pclass.map({1: "First", 2: "Second", 3: "Third"})
titanic["who"] = titanic[["age", "sex"]].apply(woman_child_or_man, axis=1)
titanic["adult_male"] = titanic.who == "man"
titanic["deck"] = titanic.cabin.str[0].map(lambda s: np.nan if s == "T" else s)
titanic["embark_town"] = titanic.embarked.map({"C": "Cherbourg", "Q": "Queenstown", "S": "Southampton"})
titanic["alive"] = titanic.survived.map({0: "no", 1: "yes"})
titanic["alone"] = ~(titanic.parch + titanic.sibsp).astype(bool)
titanic = titanic.drop(["name", "ticket", "cabin"], axis=1)
```

```python
titanic.head()
```

```python
pal = dict(man="#4682B4", woman="#CD5C5C", child="#2E8B57", male="#6495ED", female="#F08080")
```

```python
columns = titanic.select_dtypes(include=['object','bool']).columns
```

```python
columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal = OrdinalEncoder()
```

```python
values_cat = ordinal.fit_transform(titanic[columns])
```

```python
titanic[columns] = values_cat
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
x = titanic.drop(['survived','age', 'embarked', 'deck', 'embark_town'], axis=1)
y = titanic['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, X_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_regression = LogisticRegression()
```

```python

# Train the model
model_regression.fit(x_train, y_train)
```

```python
model_regression.coef_
```

```python
y_predict = model_regression.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```
