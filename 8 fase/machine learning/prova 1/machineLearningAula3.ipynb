```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler()
```

```python
fare = df['Fare'].values

# pega os valores da col Fare
```

```python
fare
```

```python
fare_scaled = scaler.fit_transform(fare.reshape(-1, 1))

# transforma os dados entre valores de -1 a 1
```

```python
fare_scaled
```

```python
df['Fare'] = fare_scaled
```

```python
df.head()
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
fare_scaled = scaler.fit_transform(fare.reshape(-1, 1))
```

```python
df['Fare'] = fare_scaled
```

```python
one_hot_encoder = preprocessing.OneHotEncoder()
```
