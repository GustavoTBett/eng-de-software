```python
from sklearn.preprocessing import OrdinalEncoder

# importa ordinal encoder
```

```python
ordinal = OrdinalEncoder()

# pega a var do ordinal encoder
```

```python
species_cat = dfPeixes['Species']
```

```python
species_cat = ordinal.fit_transform(species_cat.values.reshape(-1,1))

# transformando as espécies em números
```

```python
ordinal.categories_[0]
```

```python
encoding_dict = dict(zip(ordinal.categories_[0], range(len(ordinal.categories_[0]))))
```

```python
encoding_dict
```

```python
dfPeixes['Species'] = species_cat
```

```python
dfPeixes
```

```python
x = dfPeixes.drop(['Species'], axis=1)
```

```python
x
```

```python
y = dfPeixes['Species']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

# importando train test split
```

```python
x_train, X_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=42)

# realizando o split
```

```python
from sklearn.neighbors import KNeighborsClassifier

# importando o KNN
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3)

# pegando a var do KNN
```

```python
model_knn.fit(x_train, y_train)

# treinando o modelo
```

```python
y_predict = model_knn.predict(X_test)

# realizando a predição
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score

# importando o accuracy score
```

```python
accuracy_score(y_test, y_predict)

# calculando a acurácia
```

```python
list_result = list()
# percorrendo uma lista n vizinhos de 1 até 100 com passo de 3
for n_knn in range(1, 100, 2):
    model_knn = KNeighborsClassifier(n_neighbors=n_knn)
    model_knn.fit(x_train, y_train)
    y_predict = model_knn.predict(X_test)
    acc = accuracy_score(y_test, y_predict)
    list_result.append((acc, n_knn))
```

```python
dfPeixes_result = pd.DataFrame(list_result, columns=['accuracy', 'n_neighbors'])
```
